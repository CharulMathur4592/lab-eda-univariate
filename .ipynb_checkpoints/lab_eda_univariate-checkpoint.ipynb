```python
# Part 1: Understanding Product Categories

# 1. Frequency Tables:

    # 1.1 Generate a frequency table for the product 'category'.
frequency_table = df['category'].value_counts()
frequency_table
```

```python
# 1.2 Which are the top 5 most listed product categories?

top_5_categories = df['category'].value_counts().head(5)

print("Top 5 Most Listed product categories:\n", top_5_categories)
```

```python
# 2. Visualizations:
# 2.1 Display the distribution of products across different categories using a bar chart. 
# If you face problems understanding the chart, do it for a subset of top categories.

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category', order=top_5_categories.index)
plt.title("Distribution of products across Top 5 Product Categories")
plt.xlabel("Product Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# 2.2 For a subset of top categories, visualize their proportions using a pie chart. 

plt.figure(figsize=(8, 8))
plt.pie(top_5_categories.values, labels=top_5_categories.index, autopct='%1.1f%%', startangle=140)
plt.title(f"Proportion of Top Product Categories")
plt.tight_layout()
plt.show()

# Does any category dominate the listings?
percentages = (top_5_categories/top_5_categories.sum()) * 100
print(percentages)
```

```python
# Part 2: Delving into Product Pricing
# 1. Measures of Centrality:
# 1.1 Calculate the mean, median, and mode for the price of products.

mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print("Mean: ", mean_price, "\nMedian: ", median_price, "\nMode: ", mode_price)
```

```python
# 1.2 What's the average price point of products listed? How does this compare with the most common price point (mode)?

average_price = df['price'].mean()

print(f"Average Price (Mean): £{average_price:.2f}")
print(f"Most Common Price (Mode): £{mode_price:.2f}")
```

```python
# 2. Measures of Dispersion:

# 2.1 Determine the variance, standard deviation, range, and interquartile range for product price.

variance_price = df['price'].var()

std_dev_price = df['price'].std()

price_range = df['price'].max() - df['price'].min()

q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3 - q1

print("Variance: ", variance_price, "\nStandard Deviation: ", std_dev_price)
print("Range: ", price_range, "\nInterquartile: ", iqr)
```

```python
# 2.2 How varied are the product prices? Are there any indicators of a significant spread in prices?

product_prices = df['price']
summary = {
    "Mean": product_prices.mean(),
    "Standard Deviation": product_prices.std(),
    "Coefficient of Variation (%)": (product_prices.std() / product_prices.mean()) * 100,
    "Min Price": product_prices.min(),
    "Max Price": product_prices.max(),
    "Range": product_prices.max() - product_prices.min(),
    "IQR": product_prices.quantile(0.75) - product_prices.quantile(0.25)
}

for k, v in summary.items():
    print(f"{k}: {v:.2f}")
```

```python
# 3. Visualizations:
# 3.1 Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. 
# If its hard to read these diagrams, think why this is, and explain how it could be solved..

product_prices = df['price']

plt.figure(figsize=(8, 5))
sns.histplot(product_prices, bins=50, kde=True)
plt.xlabel("Product Price (£)")
plt.ylabel("Number of Products")
plt.title("Distribution of Product Prices")
plt.tight_layout()
plt.show()
```

```python
# 3.2 Are there products that are priced significantly higher than the rest? 
# Use a box plot to showcase the spread and potential outliers in product pricing.

# Plot box plot
plt.figure(figsize=(10, 4))
sns.boxplot(x=product_prices)
plt.xlabel("Product Price (£)")
plt.title("Box Plot of Product Prices (Outliers Highlighted)")
plt.tight_layout()
plt.show()
```

```python
# Part 3: Unpacking Product Ratings
# Business Question: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

# 1. Measures of Centrality:

# 1.1 Calculate the mean, median, and mode for the rating of products.
product_rating = df['stars']

mean_rating = product_rating.mean()
median_rating = product_rating.median()
mode_rating = product_rating.mode()[0]

print(f"Mean: {mean_rating:.2f}")
print(f"Median: {median_rating:.2f}")
print(f"Mode: {mode_rating:.2f}")
```

```python
# 1.2 How do customers generally rate products? Is there a common trend?
    # Customers generally rate products positively. Clear Trend towards high rating.
```

```python
# 2 Measures of Dispersion:

# 2.1 Determine the variance, standard deviation, and interquartile range for product rating.
variance_rating = product_rating.var()

std_dev_rating = product_rating.std()

q1 = product_rating.quantile(0.25)
q3 = product_rating.quantile(0.75)
iqr = q3 - q1

print(f"Variance: {variance_rating:.2f}")
print(f"Standard Deviation: {std_dev_rating:.2f}")
print("Interquartile: ", iqr)

# 2.2 Are the ratings consistent, or is there a wide variation in customer feedback?
    # Low variance: Ratings are tightly clustered around the mean
    # Small standard deviation: Most customers give similar ratings
    # Narrow IQR: The middle 50% of ratings fall within a small range (e.g., 4–5)
```

```python
# 3. Shape of the Distribution:

# 3.1 Calculate the skewness and kurtosis for the rating column.

skewness = product_rating.skew()
kurtosis = product_rating.kurt()   # Fisher's kurtosis (normal = 0)

print(f"Skewness: {skewness:.3f}")
print(f"Kurtosis: {kurtosis:.3f}")

# 3.2 Are the ratings normally distributed, or do they lean towards higher or lower values?
    # Product ratings are not normally distributed — they lean toward higher values.
```

```python
# 4. Visualizations:

# 4.1 Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?

plt.figure(figsize=(7, 4))
sns.histplot(product_rating, bins=10, kde=True)
plt.xlabel("Product Rating")
plt.ylabel("Number of Products")
plt.title("Distribution of Product Ratings")
plt.tight_layout()
plt.show()
```
